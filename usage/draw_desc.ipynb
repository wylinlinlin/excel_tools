```python
import pandas as pd
import pyecharts.options as opts
from pyecharts.charts import Bar

df = pd.read_excel("C:/Users/Desktop/matplotlib/demo1.xlsx")

df["desc_len"] = df["摘要"].str.len()

df["desc_len"] = df["desc_len"].fillna(0)

def bubble_sort(lst):
  count = len(lst)
  for i in range(count):
    for j in range(i + 1, count):
      if lst[i] > lst[j]:
        lst[i], lst[j] = lst[j], lst[i]
  return lst

tmp = df['desc_len'].value_counts().index.tolist()
tmp_sort = bubble_sort(tmp)

name_lst = []
cnt_lst = []

for idx, name in enumerate(tmp_sort):
  name_lst.append(name)
  cnt_lst.append(df['desc_len'].value_counts()[name])

cnt_lst = list(map(int, cnt_lst))

bar = Bar()
bar.add_xaxis(name_lst)
bar.add_yaxis("cnt", cnt_lst)
bar.set_global_opts(title_opts=opts.TitleOpts(title="摘要长度频率统计"),
                   toolbox_opts=opts.ToolboxOpts(is_show=True),
                   )
bar.set_series_opts(label_opts=opts.LabelOpts(position="top"))
bar.render_notebook() 
# bar.render("C:/Users/Desktop/matplotlib/demo_bar.html")
```
