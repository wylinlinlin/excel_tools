```python
# 合并不同文件夹相同名字的excel
import pandas as pd
import os


def merge_data(id):
    base_ = r'C:\Users\hanstan\Desktop\433383\序时账 都城1-9\10_7家分立\mer_acc'  # 大文件夹
    base = f"{base_}/"
    dir_names = os.listdir(base)  # 打开大文件夹后的各个小文件夹名dir_names
    df_all = pd.DataFrame()
    for dir in dir_names:  # 遍历每一个小文件夹
        file_path = base+dir
        file_names = os.listdir(file_path)  # 打开小文件夹后每一个excel文件的名称
        for file in file_names:  # 遍历小文件夹里的每一个文件
            file_name = file_path+'/'+file

            if file.split('.')[0] == id:
                try:
                    df = pd.read_excel(file_name, dtype=object)
                except:
                    df = pd.read_csv(file_name, dtype=object)
               # df_all = df_all.append(df)
                df_all = df_all.append(df, ignore_index=True)  # 索引重新排序
    return df_all


if __name__ == '__main__':
    base_ = r'C:\Users\hanstan\Desktop\433383\序时账 都城1-9\10_7家分立\mer_acc'
    base = f"{base_}/"
    dir_names = os.listdir(base)
    ids = set()  # set()可以看做一个列表，这里面不包含重复的元素，不可以使用索引
    for dir in dir_names:
        file_path = base + dir
        file_names = os.listdir(file_path)
        for file_name in file_names:
            id = file_name.split('.')[0]
            ids.add(id)  # 把所有id放入ids中，这样不会有重复的id存在
    for id in ids:
        df = merge_data(id)
        base = r'C:\Users\hanstan\Desktop\433383\序时账 都城1-9\10_7家分立\acc1-12'  # 存放合并后的文件路径
        id_path = f"{base}/{id}.xlsx"
        try:
            df.to_excel(f"{base}/{id}.xlsx", index=False, encoding="gbk")
            print(f"{id}.xlsx is saved.")
        except:
            df.to_csv(f"{base}/{id}.csv", index=False, encoding="utf-8")
            print(f"{id}.csv is saved.")    
```
